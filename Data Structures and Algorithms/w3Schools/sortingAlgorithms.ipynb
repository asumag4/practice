Improved Insertion Sort 

```python
my_array = [64, 34, 25, 12, 22, 11, 90, 5]

n = len(my_array) # calculate the length
for i in range(1,n): # start from the second element to the last 
    insert_index = i 
    current_value = my_array[i]
    for j in range(i-1, -1, -1): # compare elements that are already sorted to the element that is to be sorted
        # print(j)
        if my_array[j] > current_value: # if j (going backwards) if greater than current val
            my_array[j+1] = my_array[j] # shift it one position to the right
            insert_index = j # shift the insertion index too 
        else:
            break 
                            # keep looping until the condition is met, then break
    my_array[insert_index] = current_value # insert to the correct position in the sorted portion 
    print(my_array)
print("Sorted array:", my_array)
```

QuickSort Implementation

```python
def partition(array, low, high): 
    pivot = array[high]
    i = low - 1

    for j in range(low, high): 
        if array[j] <= pivot: 
            i += 1 # Expand the smaller-than-pivot section
            array[i], array[j] = array[j], array[i]

    array[i+1], array[high] = array[high], array[i+1]
    print("Partition: ")
    print(array)
    return i+1

def quicksort(array, low=0, high=None):
    if high is None:
        high = len(array) - 1 # Pick the last value 

    if low < high: 
        pivot_index = partition(array, low, high)
        quicksort(array, low, pivot_index-1)
        quicksort(array, pivot_index+1, high)

my_array = [64, 34, 25, 12, 22, 11, 90, 5]
quicksort(my_array)
print("Sorted array:", my_array)
```

# Counting Sort 

```python
def countingSort(arr):
    max_val = max(arr)
    count = [0] * (max_val + 1)

    # Go through all the elements of the array, count and pop them
    # Use a while loop so that each time an element is popped, 'arr' is updated
    while len(arr) > 0:
        num = arr.pop(0)
        count[num] += 1

    # Go through the recorded count of each element, use nested loops to go through each index, 
    # representing the value of each element & the value for each index (value) is minused as a counter
    for i in range(len(count)):
        while count[i] > 0:
            arr.append(i)
            count[i] -= 1

    return arr

unsortedArr = [4, 2, 2, 6, 3, 3, 1, 6, 5, 2, 3]
sortedArr = countingSort(unsortedArr)
print("Sorted array:", sortedArr)
```
