```python
def merge(nums1: list[int], m: int, nums2: list[int], n: int) -> None:
    """
    Do not return anything, modify nums1 in-place instead.
    """
    # Base Case (if possible)
    # Base Case 1. If m = 0, return nums2
    if (m == 0):
        for i in range(n):
            nums1[i] = nums2[i]
        return nums1 
    # Base Case 2. If n = 0, return nums1
    elif (n == 0):
        return nums1
    else: 
        # First, remove the trailing zeros in nums1 
        del nums1[m:]
        
        # Merge Then Sort:
        nums1.extend(nums2)

        # Sometimes it might not be sorted, so sort it. 
        n = len(nums1)
        for i in range(n-1):
            swapped = False
            for j in range(n-i-1):
                if nums1[j] > nums1[j+1]:
                    nums1[j], nums1[j+1] = nums1[j+1], nums1[j]
                    swapped = True
            if not swapped:
                break

        return nums1
```
