```python
import pandas as pd
```

```python
df = pd.DataFrame(data=[
    [1, "b1", 100],
    [1, "b1", 100],
    [1, "b1", 100],
    [2, "b2", 50],
    [2, "b2", 100],
])

df.columns = ['emp_id','branch_code','sales']
df
```

```python
def topEmp(df):
    # Write your code here
    df_byEmployee = df.groupby('emp_id').sum('sales').reset_index()
    idx = df_byEmployee['sales'].idxmax()
    table = df_byEmployee.iloc[[idx,]].drop(columns=['sales'])
    return table

topEmp(df)
```
